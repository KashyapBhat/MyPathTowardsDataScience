```python
#Joins and Concatenation
```

```python
d1 = pd.DataFrame({'CustID': ['101', '102', '103', '104'],
                        'Sales': [13456, 45321, 54385, 53212],
                        'Priority': ['CAT0', 'CAT1', 'CAT2', 'CAT3'],
                        'Prime': ['yes', 'no', 'no', 'yes']},
                        index=[0, 1, 2, 3])
```

```python
d2 = pd.DataFrame({'CustID': ['101', '103', '104', '105'],
                        'Sales': [13456, 54385, 53212, 4534],
                        'Payback': ['CAT4', 'CAT5', 'CAT6', 'CAT7'],
                        'Imp': ['yes', 'no', 'no', 'no']},
                         index=[4, 5, 6, 7]) 
```

```python
d3 = pd.DataFrame({'CustID': ['101', '104', '105', '106'],
                        'Sales': [13456, 53212, 4534, 3241],
                        'Pol': ['CAT8', 'CAT9', 'CAT10', 'CAT11'],
                        'Level': ['yes', 'no', 'no', 'yes']},
                        index=[8, 9, 10, 11])
```

```python
d1
```

```python
d2
```

```python
d3
```

```python
ij = pd.merge(d1,d2)
```

```python
ij
```

```python
fj= pd.merge(d1,d2,how='outer')
```

```python
fj
```

```python
lj= pd.merge(d1,d2,how='left')
```

```python
lj
```

```python
rj= pd.merge(d1,d2,how='right')
```

```python
rj
```

```python
hc=pd.concat([d1,d2])
```

```python
hc
```

```python
vc=pd.concat([d1,d2],axis=1)
```

```python
vc
```
